```python
import numpy as np # linear algebra
import pandas as pd # data processing
import seaborn as sns # data visualization

import json # to read the JSON file from Yelp

# ignore filter warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
filename = "yelp_dataset/yelp_academic_dataset_business.json" 
df = pd.read_json(filename, lines=True) # import the .json file containing businesses data
```

```python
print(f"The restaurants dataset contains {df.shape[0]} restaurants.")
df.head(5)
```

```python
df.categories.head(20) # there are not only restaurants in the dataset
```

```python
# We will make a list of all indices at which there is a business whose category does not
# contain the word "restaurant" and then we will drop those indices

substring = "Restaurant"
not_a_restaurant = []

for i in range (0, df.shape[0]):

    fullstring = json.dumps(df.categories[i])

    if substring not in fullstring :
        not_a_restaurant.append(i)

df_restaurants = df.drop(not_a_restaurant, axis=0).reset_index()
```

```python
print(f"The dataset contains {df.shape[0]} businesses of which {df_restaurants.shape[0]} are restaurants.")
```

```python
df_restaurants.groupby(by='categories').count().sort_values(by='index', ascending=False).head(20)
```

```python
# Take a sample of 10 000 restaurants

df_restaurants_sample = df_restaurants.sample(n=10000, random_state=1, replace=False).reset_index()
```

```python
df_restaurants_sample.head(5)
```

```python
# We keep only the reviews for our previous sample of 10 000 restaurants

reviews_for_sample_of_restaurants = df_reviews[df_reviews['business_id'].isin(df_restaurants_sample['business_id'])]
```
